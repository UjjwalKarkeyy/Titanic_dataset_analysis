```python
# Meaning of the columns:

# Pclass: Passenger class, so if they were from 1st class, 2nd class, or etc. Class gives a measure for the economic value of the seat

# SibSp: The number of siblings or spouse that the passenger had aboard

# Parch: The number of children or parent the passenger had aboard

# Cabin: If the passenger had any cabin, which cabin number were they seated at

# Embarked: Which gate did they use to aboard the ship
```

```python
data = temp_data
data.tail()
```

```python
data['Sex'] = [ i.capitalize() for i in data['Sex']]
```

```python
data.iloc[10:20]
```

```python
data['Age'] = [ 0 if math.isnan(i) else i for i in data['Age']]
```

```python
data['Age'] = [math.floor(i) for i in data['Age']]
```

```python
data.tail()
```

```python
data.Ticket.unique()
```

```python
data['Ticket'] = [ i.strip() for i in data['Ticket']]
```

```python
data['Ticket'] = [ i.split(" ")[1] if " " in i else i for i in data['Ticket']]
```

```python
data.tail(20) 
```
